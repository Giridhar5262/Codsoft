```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df_train=pd.read_csv("/content/fraudTrain.csv")
df_test=pd.read_csv("/content/fraudTest.csv")
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
df_train=df_train.drop("Unnamed: 0",axis=1)
df_test=df_test.drop("Unnamed: 0",axis=1)
```

```python
df=pd.concat([df_train, df_test])
df
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
df["gender"]=label_encoder.fit_transform(df["gender"])
```

```python
col=["cc_num","gender","trans_num","unix_time","merch_lat","merch_long"]
trans_num=df["trans_num"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=500)
trans_num = vectorizer.fit_transform(trans_num)
```

```python
df["trans_num"]=trans_num.shape[0]
```

```python
x=df[col]
y=df["is_fraud"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33, random_state=42)
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
y_pred=log.predict(x_test)
accuracy=accuracy_score(y_pred,y_test)
print("Accuracy: ",accuracy)
```
