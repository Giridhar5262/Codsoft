```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = pd.read_csv('/content/spam.csv',encoding = 'latin-1')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
data.describe()
```

```python
data=data.rename(columns={"v1":"label","v2":"message"})
data.head()
```

```python
data=data.rename(columns={"Unnamed: 2":"v3","Unnamed: 3":"v4","Unnamed: 4":"v5"})
```

```python
data.columns
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.drop(["v3","v4","v5"],axis=1)
data
```

```python
data['label']
```

```python
data['message']
```

```python
data['message'] = data['message'].str.lower()
data['message']
```

```python
import re
pattern = r'[^a-zA-Z0-9\s]|\.'
data['message'] = data['message'].str.replace(pattern, '', regex=True)
```

```python
data
```

```python
data.describe()
```

```python
x = data['message']
y = data['label']
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=30)
```

```python
vectorize = TfidfVectorizer(stop_words='english')
```

```python
x_train =vectorize.fit_transform(x_train)
x_test = vectorize.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
```

```python
naive_bayes = MultinomialNB()
naive_bayes.fit(x_train, y_train)
nb = naive_bayes.predict(x_test)
accuracy_nb = accuracy_score(y_test, nb)
print("Accuracy:", accuracy_nb)
```

```python
log = LogisticRegression()
log.fit(x_train, y_train)
predictions = log.predict(x_test)
accuracy = accuracy_score(y_test,predictions)
print("Accuracy:", accuracy)
```
